```python
import pandas as pd
# Raw GitHub URL
url = "/Users/macminiadi/Documents/Programs/VSCode/Healthcare/data/NDC_Product.csv"
# Load CSV file
df_NDCProduct = pd.read_csv(url) #, delimiter="\t")
# Preview
df_NDCProduct.info()
```

```python
#check if PRODUCTID has duplicates
if df_NDCProduct['PRODUCTID'].duplicated().any():
    print(" PRODUCTID has duplicates")
else:
    print(" PRODUCTID has no duplicates")

# show distinct values in PRODUCTNDC
print("\n Distinct PRODUCTNDC values:", (df_NDCProduct['PRODUCTNDC'].nunique()))

#show distinct values in DOSAGEFORM
print("\n Distinct DOSAGEFORMNAME values:", (df_NDCProduct['DOSAGEFORMNAME'].nunique()))

#show distinct values in DOSAGEFORM
print("\n Distinct ROUTENAME values:", (df_NDCProduct['ROUTENAME'].nunique()))

```

```python
# Show distinct values of MARKETINGCATEGORYNAME
print("\n Distinct MARKETINGCATEGORYNAME values:", (df_NDCProduct['MARKETINGCATEGORYNAME'].nunique()))

print("\nUnique MARKETING CATEGORY NAME values are:")
for i, category in enumerate(df_NDCProduct['MARKETINGCATEGORYNAME'].dropna().unique(), start=1):
    print(f"{i}. {category}")

#Show Marketing category names that contains 'unapproved'
print("\n Marketing categories that contain 'unapproved':")
unapproved_categories = df_NDCProduct[df_NDCProduct['MARKETINGCATEGORYNAME'].str.contains('unapproved', case=False, na=False)]
if not unapproved_categories.empty:
    for i, category in enumerate(unapproved_categories['MARKETINGCATEGORYNAME'].unique(), start=1):
        print(f"{i}. {category}")
  

#show ENDMARKETINGDATE 
print("\n Distinct EENDMARKETINGDATE  values:", (df_NDCProduct['ENDMARKETINGDATE'].nunique()))

#format ENDMARKETINGDATE to datetime
df_NDCProduct['ENDMARKETINGDATE'] = pd.to_datetime(df_NDCProduct['ENDMARKETINGDATE'], errors='coerce')
df_NDCProduct['STARTMARKETINGDATE'] = pd.to_datetime(df_NDCProduct['STARTMARKETINGDATE'], errors='coerce')

#Show ENDMARKETINGDATE in ascending order
print("\n ENDMARKETINGDATE in descending order:")
print(df_NDCProduct['ENDMARKETINGDATE'].sort_values(ascending=False).head(5))
print(df_NDCProduct['STARTMARKETINGDATE'].sort_values(ascending=False).head(5))
```
